```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
yt = pd.read_csv("youtubers.csv").drop(yt.columns[0], axis = 1)
yt
```

## EDA

```python
totals = yt.groupby('sex').mean().reset_index()
totals
```

```python
np.arange(len(totals.iloc[0]))
```

```python
plt.bar(totals['sex'], totals['sub_count'],color = 'blue')
plt.xlabel('Sex')
plt.ylabel('Average Subscriber Count')
plt.title("Average Sub Count Bewteen M and F Youtubers")
```

```python
plt.bar(totals['sex'], totals['num_views'],color = 'pink')
plt.xlabel('Sex')
plt.ylabel('Average View Count')
plt.title("Average View Count Bewteen M and F Youtubers")
```

```python
plt.bar(totals['sex'], totals['num_vids'],color = 'gray')
plt.xlabel('Sex')
plt.ylabel('Average Video Count')
plt.title("Average Video Count Bewteen M and F Youtubers")
```

```python
#proportions views : vid
yt['avg_views_per_vid'] = yt['num_views']/yt['num_vids']
```

```python
totals = yt.groupby('sex').mean().reset_index()
```

```python
plt.bar(totals["sex"], totals['avg_views_per_vid'])
plt.xlabel('Sex')
plt.ylabel('Average View per Video Count')
plt.title("Average View per Video Count Bewteen M and F Youtubers")
```

```python
# proportion subs to views
yt['sub_to_views_props'] = yt['sub_count']/yt['num_views']
```

```python
totals = yt.groupby('sex').mean().reset_index()
```

```python
plt.bar(totals["sex"], totals['sub_to_views_props'])
plt.xlabel('Sex')
plt.ylabel('Sub to Views Proportion')
plt.title("Sub to Views Proportion Bewteen M and F Youtubers")
```

```python
yt['sub_count'].plot(kind = 'hist', 
                     x = 'sub_count',
                    xlabel = "Subscriber Count",
                    title = "Subscriber Count Histogram; All")
```

```python
yt[yt.get("sex")=="F"]['sub_count'].plot(kind = 'hist', 
                                         x = 'sub_count',
                                        xlabel = "Subscriber Count",
                                        title = "Subscriber Count Histogram",
                                        legend = True)

yt[yt.get("sex")=="M"]['sub_count'].plot(kind = 'hist', 
                                         x = 'sub_count',
                                        xlabel = "Subscriber Count",
                                        title = "Subscriber Count Histogram",
                                        alpha = 0.7,
                                        legend = True)
plt.legend(["F sub count", 'M sub count'])
plt.xlabel("Subscriber Count")
```

```python
yt[yt.get("sex")=="F"]['num_vids'].plot(kind = 'hist', 
                                         x = 'num_vids',
                                        xlabel = "Video Count",
                                        title = "Video Count Histogram",
                                        legend = True)

yt[yt.get("sex")=="M"]['num_vids'].plot(kind = 'hist', 
                                         x = 'num_vids',
                                        xlabel = "Video Count",
                                        title = "Video Count Histogram",
                                        alpha = 0.7,
                                        legend = True)
plt.legend(["F vid count", 'M vid count'])
plt.xlabel("Video Count")
```

```python
yt[yt.get("sex")=="F"]['num_views'].plot(kind = 'hist', 
                                         x = 'num_views',
                                        xlabel = "Video Count",
                                        title = "Video Count Histogram",
                                        legend = True)

yt[yt.get("sex")=="M"]['num_views'].plot(kind = 'hist', 
                                         x = 'num_views',
                                        xlabel = "View Count",
                                        title = "View Count Histogram",
                                        alpha = 0.7,
                                        legend = True)
plt.legend(["F view count", 'M view count'])
plt.xlabel("View Count")
```

```python
yt[yt.get('sub_to_views_props')==yt['sub_to_views_props'].max()]
```

```python
yt[yt.get("sex")=="F"]['sub_to_views_props'].plot(kind = 'hist', 
                                         x = 'sub_to_views_props',
                                        xlabel = "Sub:Views Ratio",
                                        title = "Sub to Views Ratio Histogram",
                                        legend = True)

yt[yt.get("sex")=="M"]['sub_to_views_props'].plot(kind = 'hist', 
                                         x = 'sub_to_views_props',
                                        xlabel = "Sub:Views Ratio",
                                        title = "Sub to Views Ratio Histogram",
                                        alpha = 0.7,
                                        legend = True)
plt.legend(["F ratio", 'M ratio'])
plt.xlabel("Sub:Views")
```

```python
yt[yt.get("sex")=="F"]['avg_views_per_vid'].plot(kind = 'hist', 
                                         x = 'avg_views_per_vid',
                                        xlabel = "Avg View per Video",
                                        title = "Avg View per Video",
                                        legend = True)

yt[yt.get("sex")=="M"]['avg_views_per_vid'].plot(kind = 'hist', 
                                         x = 'avg_views_per_vid',
                                        xlabel = "Avg View per Video",
                                        title = "Avg View per Video",
                                        alpha = 0.7,
                                        legend = True)
plt.legend(["F avg view per video", 'M avg view per video'])
plt.xlabel("Avg View per Video")
```
